```python
def fd(CrCbIm, mean, Std):
    gauss2D = multivariate_normal(mean,Std)
    Skin = (gauss2D.pdf(CrCbIm) > 0.25).astype('uint8')
    
    kernel_small = np.ones((3, 3), np.uint8)
    kernel_big = np.ones((13, 13), np.uint8)
    
    plt.imshow(Skin)
    plt.show()
    
    Skin = cv2.morphologyEx(Skin, cv2.MORPH_OPEN, kernel_small, iterations=1) # Opening
    
    Skin = cv2.morphologyEx(Skin, cv2.MORPH_CLOSE, kernel_big, iterations=1) # Closing
    
    Skin, num_features = label(Skin)
    
    plt.imshow(Skin)
    plt.show()
    
    Bounding_Boxes = []
    for i in range(1,num_features + 1):
        Bounding_Box = []
        Area = (Skin == i)
        Indeces = np.nonzero(Area)
        Bounding_Box.append(min(Indeces[1]))
        Bounding_Box.append(min(Indeces[0]))
        Bounding_Box.append(max(Indeces[1]) - min(Indeces[1]))
        Bounding_Box.append(max(Indeces[0]) - min(Indeces[0]))
        Bounding_Boxes.append(Bounding_Box)
        
    return Bounding_Boxes
    
```

```python
def Gaussian_Filters_v2(sigma): # from lab1
    n = int(np.ceil(3 * sigma) * 2 + 1)
    Gaussian_1D = cv2.getGaussianKernel(n, sigma)
    Gaussian = Gaussian_1D @ Gaussian_1D.T
    return Gaussian

def lk(I1, I2, features, rho, epsilon, d_x0, d_y0): # Lucas-Kanade algorithm implementation
    
    grad_y, grad_x = np.gradient(I1)
    
    gaussian = Gaussian_Filters_v2(rho)
    
    dx_final = d_x0
    dy_final = d_y0
    
    height, width = I1.shape
    x_0, y_0 = np.meshgrid(np.arange(width), np.arange(height))
    
    for i in range(features.shape[0]):
        
        x = features[i][0][0]
        y = features[i][0][1]
        dx = d_x0[y][x]
        dy = d_y0[y][x]
        u = np.array([[100.0],        
                      [100.0]]) # initializing u with a high value (2x1 vector)
        
        tolerance = 1e-3  # the tolerance level for convergence
        
        prev_dx = 10 # It will enter the while loop
        prev_dy = 10
        
        while (np.abs(dx - prev_dx) > tolerance) or (np.abs(dy - prev_dy) > tolerance): # criterion of convergence
            
            prev_dx = dx
            prev_dy = dy
            
            I1_d = map_coordinates(I1, [y_0 + dy, x_0 + dx], order=1, mode='nearest').reshape(I1.shape)
            a2 = map_coordinates(grad_y, [y_0 + dy, x_0 + dx], order=1, mode='nearest').reshape(I1.shape)
            a1 = map_coordinates(grad_x, [y_0 + dy, x_0+ dx], order=1, mode='nearest').reshape(I1.shape)
            
            e = I2 - I1_d
            
            inv11 = cv2.filter2D(a1**2, -1, gaussian) + epsilon
            inv21 = cv2.filter2D(a1 * a2, -1, gaussian)
            inv12 = inv21
            inv22 = cv2.filter2D(a2**2, -1, gaussian) + epsilon
            
            invArray = np.array([[inv11[y, x], inv12[y, x]],
                                 [inv21[y, x], inv22[y, x]]])
            
            newArray = np.linalg.inv(invArray) # computing the inverse of invArray
            
            norm1 = cv2.filter2D(a1 * e, -1, gaussian)
            norm2 = cv2.filter2D(a2 * e, -1, gaussian)
            
            normArray = np.array([[norm1[y, x]],
                                  [norm2[y, x]]])
            
            u = newArray @ normArray

            dy = dy + u[1, 0]
            dx = dx + u[0, 0]

        dx_final[y][x] = -dx
        dy_final[y][x] = -dy
    
    return dx_final, dy_final
```

```python
def optical_flow_tvL1(box_1, box_2): # we use this function to output optical flow in a more convenient way
    
    optical_flow = cv2.DualTVL1OpticalFlow_create()
    flow = optical_flow.calc(box_1, box_2, None)
    
    flow_x = flow[..., 0]
    flow_y = flow[..., 1]
    
    x_components = flow_x[..., 0] 
    y_components = flow_y[..., 1]

    return np.column_stack((x_components, y_components)).tolist()
```

```python
def displ_lk(dx,dy): 
    
    sum_of_squares = np.square(dx) + np.square(dy)
    
    winners = sum_of_squares > 0.5*(np.sum(sum_of_squares)/len(np.nonzero(sum_of_squares)[0])) 
    # the average value is used as the threshold
    
    masked_sum_of_dx = np.where(winners, dx, np.zeros(winners.shape))
    masked_sum_of_dy = np.where(winners, dy, np.zeros(winners.shape))
    
    return ((np.sum(masked_sum_of_dx)/len(np.nonzero(masked_sum_of_dx)[0])) , (np.sum(masked_sum_of_dy)/len(np.nonzero(masked_sum_of_dy)[0])) )   
```

```python
def displ_tvl1(d): 
    
    dx, dy = zip(*d)
    sum_of_squares = np.square(dx) + np.square(dy)
    
    winners = sum_of_squares > np.mean(sum_of_squares) # the average value is used as the threshold
    
    masked_sum_of_dx = np.ma.masked_array(dx, mask=~winners)
    masked_sum_of_dy = np.ma.masked_array(dy, mask=~winners)
    
    return (np.mean(masked_sum_of_dx), np.mean(masked_sum_of_dy))   
```

```python
# Lowpass filter

def Lowpass_Filter(a): # a = 0.6 is good for gauss pyramid generation
    k = np.zeros((5,1))
    k[2] = a
    k[1] = k[3] = 1/4
    k[0] = k[4] = 1/4 - a/2
    return k @ k.T
```

```python
def GREDUCE(Im,h):
    
    # Performing Lowpass filtering with kernel h and decimation 
    Im_filtered = cv2.filter2D(Im, -1, h)
    
    g = cv2.pyrDown(Im_filtered)
    
    return g
```

```python
def GPyramid(Im, a, depth):
    h = Lowpass_Filter(a)
    G = []

    G.append(Im) # G[0] is the original image

    for level in range(depth):
        G.append(GREDUCE(G[level],h))

    return G
```

```python
def multiscale_lk(I1, I2, rho, epsilon, depth):
    # Generate Pyramids
    G1 = GPyramid(I1, 0.6, depth) # Gaussian pyramid of first frame
    G2 = GPyramid(I2, 0.6, depth) # Gaussian pyramid of following frame
    
    corners = cv2.goodFeaturesToTrack(I1, 1000, 0.1, 0.1) 
    corners = np.int0(corners//(2**depth))
    
    dx = np.zeros(G1[-1].shape) # init
    dy = np.zeros(G1[-1].shape)
    for i in range(depth):
        I1 = G1.pop()
        I2 = G2.pop()
        dx, dy = lk(I1, I2, corners, rho, epsilon, dx,  dy)
        dx_new = np.zeros(G1[-1].shape) # init
        dy_new = np.zeros(G1[-1].shape)
        corners_new = np.int0(2*corners)
        for j in range(corners.shape[0]):
            dx_new[corners_new[j][0][1]][corners_new[j][0][0]] = 2*dx[corners[j][0][1]][corners[j][0][0]]
            dy_new[corners_new[j][0][1]][corners_new[j][0][0]] = 2*dy[corners[j][0][1]][corners[j][0][0]]
        dx = dx_new
        dy = dy_new
        corners = corners_new
    
    I1 = G1.pop()
    I2 = G2.pop()
    dx, dy = lk(I1, I2, corners, rho, epsilon, dx,  dy)
        
    return dx, dy
```

```python
## Harris-stephens detector 3D (x,y,t)

def Harris_Stephens(V, sigma, tau, s, theta_corn, k):
    '''
    Ανιχνευτής γωνιών Harris-Stephens που υλοποιήθηκε στο 1ο εργαστήριο απλά για βίντεο. Η υλοποίηση του περιγράφεται
    αναλυτικά στην εκφώνηση. Επιστρέφει τις 500 μεγαλύτερες τιμές του κριτηρίου γωνιότητας. Σχιλιασμό: Αυτό έχει αποτέλεσμα
    πολλές μαζεμένες γωνίες --> Μεγάλη συσχέτιση --> ανούσιο και ίσως κακό για το classification.
    
    '''
    n1 = int(np.ceil(3 * sigma) * 2 + 1)
    Gaussian_sigma = cv2.getGaussianKernel(n1, sigma)
    Gaussian_sigma = Gaussian_sigma.reshape(Gaussian_sigma.shape[0])
    
    n2 = int(np.ceil(3 * tau) * 2 + 1)
    Gaussian_tau = cv2.getGaussianKernel(n2, tau)
    Gaussian_tau = Gaussian_tau.reshape(Gaussian_tau.shape[0])
    
    I_gaussblur_3D = convolve1d(convolve1d(convolve1d(V,Gaussian_sigma,axis=0),Gaussian_sigma,axis=1),Gaussian_tau,axis=2)
    
    Ly, Lx, Lt = np.gradient(I_gaussblur_3D)
    
    n1 = int(np.ceil(3 * s*sigma) * 2 + 1)
    Gaussian_sigma_s = cv2.getGaussianKernel(n1, s*sigma)
    Gaussian_sigma_s = Gaussian_sigma_s.reshape(Gaussian_sigma_s.shape[0])
    
    n2 = int(np.ceil(3 * s*tau) * 2 + 1)
    Gaussian_tau_s = cv2.getGaussianKernel(n2, s*tau)
    Gaussian_tau_s = Gaussian_tau_s.reshape(Gaussian_tau_s.shape[0])
    
    M_1_1 = convolve1d(convolve1d(convolve1d(Lx**2,Gaussian_sigma_s,axis=0),Gaussian_sigma_s,axis=1),Gaussian_tau_s,axis=2)
    M_2_2 = convolve1d(convolve1d(convolve1d(Ly**2,Gaussian_sigma_s,axis=0),Gaussian_sigma_s,axis=1),Gaussian_tau_s,axis=2)
    M_3_3 = convolve1d(convolve1d(convolve1d(Lt**2,Gaussian_sigma_s,axis=0),Gaussian_sigma_s,axis=1),Gaussian_tau_s,axis=2)
    M_2_1 = convolve1d(convolve1d(convolve1d(Lx*Ly,Gaussian_sigma_s,axis=0),Gaussian_sigma_s,axis=1),Gaussian_tau_s,axis=2)
    M_3_1 = convolve1d(convolve1d(convolve1d(Lx*Lt,Gaussian_sigma_s,axis=0),Gaussian_sigma_s,axis=1),Gaussian_tau_s,axis=2)
    M_3_2 = convolve1d(convolve1d(convolve1d(Ly*Lt,Gaussian_sigma_s,axis=0),Gaussian_sigma_s,axis=1),Gaussian_tau_s,axis=2)
    
    trace = M_1_1 + M_2_2 + M_3_3
    det = M_1_1*M_2_2*M_3_3 + 2*M_2_1*M_3_1*M_3_2 - M_1_1*M_3_2**2 - M_2_2*M_3_1**2 - M_3_3*M_2_1**2

    H = det - k*trace**3
    
    H = H[sigma:-sigma,sigma:-sigma,:]
    
    # plt.imshow(H[:,:,66])
    # plt.show()
    
    # plt.imshow(H[:,:,86])
    # plt.show()
    
    # Synthikh = H > theta_corn*np.maximum(H) !!! not needed
    
    ind_y, ind_x, ind_t = np.unravel_index(np.argsort(-1*H, axis=None), H.shape)
    
    scale = sigma*np.ones(ind_x.shape)
    
    Interest_points = list(zip(ind_x,ind_y,ind_t,scale))
    
    return np.asarray(Interest_points[:500])
    
    
Interest_points = Harris_Stephens(V, 4, 1.5, 2, 0.05, 0.005)
```

```python
## Gabor detector 3D (x,y,t)

def Gabor(V, sigma, tau, theta_corn):
    '''
    Ανιχνευτής σημείων ενδιαφέροντος Gabor. Η υλοποίηση του περιγράφεται
    αναλυτικά στην εκφώνηση. Επιστρέφει τις 500 μεγαλύτερες τιμές του κριτηρίου γωνιότητας. Σχιλιασμό: Αυτό έχει αποτέλεσμα
    πολλές μαζεμένες γωνίες --> Μεγάλη συσχέτιση --> ανούσιο και ίσως κακό για το classification.
    
    '''
    n1 = int(np.ceil(3 * sigma) * 2 + 1)
    Gaussian_sigma = cv2.getGaussianKernel(n1, sigma)
    Gaussian_sigma = Gaussian_sigma.reshape(Gaussian_sigma.shape[0])
    
    omega = 4.0/tau
    t = np.linspace(-2*tau,2*tau,int(4*tau+1))
    
    Gabor_tau_ev = (np.cos(2*np.pi*t*omega)*np.exp(-t**2 / (2 * tau**2)))        
    Gabor_tau_od = (np.sin(2*np.pi*t*omega)*np.exp(-t**2 / (2 * tau**2)))
    
    Gabor_tau_ev = Gabor_tau_ev/np.linalg.norm(Gabor_tau_ev, ord=1)
    Gabor_tau_od = Gabor_tau_od/np.linalg.norm(Gabor_tau_od, ord=1)
    
    I_ev = convolve1d(convolve1d(convolve1d(V,Gaussian_sigma,axis=0),Gaussian_sigma,axis=1),Gabor_tau_ev,axis=2)
    
    I_od = convolve1d(convolve1d(convolve1d(V,Gaussian_sigma,axis=0),Gaussian_sigma,axis=1),Gabor_tau_od,axis=2)
    
    H = I_ev**2 + I_od**2
    
    H = H[sigma:-sigma,sigma:-sigma,:]
    
    # plt.imshow(H[:,:,188])
    # plt.show()
    
    # plt.imshow(H[:,:,189])
    # plt.show()
    
    # Synthikh = H > theta_corn*np.maximum(H) !!! not needed
    
    ind_y, ind_x, ind_t = np.unravel_index(np.argsort(-1*H, axis=None), H.shape)
    
    scale = sigma*np.ones(ind_x.shape)
    
    Interest_points = list(zip(ind_x,ind_y,ind_t,scale))
    
    return np.asarray(Interest_points[:500])
    
    
Interest_points = Gabor(V, 4, 1.5, 0.05)
```

```python
def out_of_bounds(feature):
    sigma = feature[3]
    return ((feature[1] - sigma//2) < 0) or ((feature[1] + sigma//2) > 120) or ((feature[0] - sigma//2) < 0) or ((feature[0] + sigma//2) > 160)
```

```python
# Time-spacial histogram descriptors

def hist_descriptor(V, features, descriptor, nbins):
    '''
    Επιστρέφει τους περιγραφητές των γωνιών. Χρησιμοποιεί σαν ιστόγραμμα είτε gradient είτε flow --> orientation_histogram(). 
    Με HOG/HOF απλά επιστρέφει και τους δύο (2*500)
    
    '''
    histdescription = []
    if (descriptor == 'HOG') or (descriptor == 'HOG/HOF'):
        
        Vy, Vx, Vt = np.gradient(V)
        
        for feature in list(features):
            # cut sigma * sigma
            if out_of_bounds(feature):
                continue
            else:
                sigma = int(feature[3])
                Gx = Vx[int(feature[1]-sigma//2):int(feature[1]+sigma//2),int(feature[0]-sigma//2):int(feature[0]+sigma//2),int(feature[2])]
                Gy = Vy[int(feature[1]-sigma//2):int(feature[1]+sigma//2),int(feature[0]-sigma//2):int(feature[0]+sigma//2),int(feature[2])]
                
                desc = p3.orientation_histogram(Gx,Gy,nbins,np.array([sigma,sigma]))
                histdescription.append(desc)
        
    if (descriptor == 'HOF') or (descriptor == 'HOG/HOF'):
        
        Flow_x = []
        Flow_y = []
        for i in range(V.shape[2] - 1):
            
            temp = V[:,:,i]*241
            epomeno = V[:,:,i+1]*241
            optical_flow = cv2.DualTVL1OpticalFlow_create()
            flow = optical_flow.calc(temp.astype('uint8'), epomeno.astype('uint8'), None)
                        
            Flow_x.append(flow[:,:,1])
            Flow_y.append(flow[:,:,0])
            
        for feature in list(features):
            # cut sigma * sigma
            if out_of_bounds(feature):
                continue
            else:
                if int(feature[2]) == 199:
                    continue
                else:
                    sigma = int(feature[3])
                    Gx = Flow_x[int(feature[2])][int(feature[1]-sigma//2):int(feature[1]+sigma//2),int(feature[0]-sigma//2):int(feature[0]+sigma//2)]
                    Gy = Flow_y[int(feature[2])][int(feature[1]-sigma//2):int(feature[1]+sigma//2),int(feature[0]-sigma//2):int(feature[0]+sigma//2)]
                
                    desc = p3.orientation_histogram(Gx,Gy,nbins,np.array([sigma,sigma]))
                    histdescription.append(desc)
        
    return np.asarray(histdescription)    
        
        
# HOG = hist_descriptor(V, Interest_points, 'HOG', 8)
# HOF = hist_descriptor(V, Interest_points, 'HOF', 8)
# HOG_HOF = hist_descriptor(V, Interest_points, 'HOG/HOF', 8)
```

```python
'''
Φορτώνω test-train (train = τα υπόλοιπα βίντεο από αυτά που μας έδωσαν)

'''

training_set = []
testing_set = []

i = 0
with open('./cv23_lab2_material/part2 - SpatioTemporal/traininng_videos.txt') as file:
        # reading each line   
        for video in file:
            if (i//12 == 0):
                training_set.append(p3.read_video('./cv23_lab2_material/part2 - SpatioTemporal/running/{}'.format(video),200,0))
            elif (i//12 == 1):
                training_set.append(p3.read_video('./cv23_lab2_material/part2 - SpatioTemporal/handwaving/{}'.format(video),200,0))
            elif (i//12 == 2):
                training_set.append(p3.read_video('./cv23_lab2_material/part2 - SpatioTemporal/walking/{}'.format(video),200,0))
            i += 1
        
i = 0
with open('./cv23_lab2_material/part2 - SpatioTemporal/testing_videos.txt') as file:
        # reading each line   
        for video in file:
            if (i//4 == 0):
                testing_set.append(p3.read_video('./cv23_lab2_material/part2 - SpatioTemporal/running/{}'.format(video),200,0))
            elif (i//4 == 1):
                testing_set.append(p3.read_video('./cv23_lab2_material/part2 - SpatioTemporal/handwaving/{}'.format(video),200,0))
            elif (i//4 == 2):
                testing_set.append(p3.read_video('./cv23_lab2_material/part2 - SpatioTemporal/walking/{}'.format(video),200,0))
            i += 1

        
print(len(training_set))
print(len(testing_set))

print(training_set[0].shape)
print(testing_set[0].shape)

# Normalize to [0,1]
for i in range(len(training_set)):
    training_set[i] = training_set[i].astype(float)/np.max(training_set[i])

for i in range(len(testing_set)):
    testing_set[i] = testing_set[i].astype(float)/np.max(testing_set[i])
    
print("Range: %f - %f " % (np.min(training_set[0]), np.max(training_set[0])))
print("Range: %f - %f " % (np.min(testing_set[0]), np.max(testing_set[0])))
```

```python
'''
Ένα παράδειγμα classification: εξηγώ παρακάτω
'''

import warnings
warnings.filterwarnings("ignore")

import cv23_lab1_part3_utils as p1
from tqdm import tqdm

training_set_desc = []
testing_set_desc = []

for V in tqdm(training_set):
    Interest_points = Harris_Stephens(V, 4, 1.5, 2, 0.05, 0.005)
    training_set_desc.append(hist_descriptor(V, Interest_points, 'HOG', 8))
    
for V in tqdm(testing_set):
    Interest_points = Harris_Stephens(V, 4, 1.5, 2, 0.05, 0.005)
    testing_set_desc.append(hist_descriptor(V, Interest_points, 'HOG', 8))
 
bow_train, bow_test = p3.bag_of_words(np.asarray(training_set_desc), np.asarray(testing_set_desc), 64)
```

```python
# Train an SVM to classify videos

train_labels = list(np.zeros(12)) + list(np.ones(12)) + list(2*np.ones(12))
test_labels = list(np.zeros(4)) + list(np.ones(4)) + list(2*np.ones(4))

accuracy, pred = p3.svm_train_test(bow_train, train_labels, bow_test, test_labels)

print("Interest point detector: Harris-Stephens and Descriptor: HOG achieve accuracy: " + str(round(accuracy,2)))

print(test_labels)
print(pred)
```

```python
'''
Τους descriptors (επιλογή ανιχνευτή + περιγραφητή) τους δίνω στην bag_of_words() αυτή βγάζει ένα feature vector για κάθε βίντεο 
και μετά εκπαιδεύω ένα SVM() για classification (ας επαιρνες Μηχ. Μαθ.) και κάνω predict το test set.

'''

import warnings
warnings.filterwarnings("ignore")

import cv23_lab1_part3_utils as p1
from tqdm import tqdm

Detectors = ['Harris-Stephens', 'Gabor']
Descriptors = ['HOG', 'HOF', 'HOG/HOF']

for detector in Detectors:
    for descriptor in Descriptors:
        training_set_desc = []
        testing_set_desc = []
        if detector == 'Harris-Stephens':
                for V in tqdm(training_set):
                    Interest_points = Harris_Stephens(V, 4, 1.5, 2, 0.05, 0.005)
                    training_set_desc.append(hist_descriptor(V, Interest_points, descriptor, 8))
    
                for V in tqdm(testing_set):
                    Interest_points = Harris_Stephens(V, 4, 1.5, 2, 0.05, 0.005)
                    testing_set_desc.append(hist_descriptor(V, Interest_points, descriptor, 8))
        else:
                for V in tqdm(training_set):
                    Interest_points = Gabor(V, 4, 1.5, 0.05)
                    training_set_desc.append(hist_descriptor(V, Interest_points, descriptor, 8))
    
                for V in tqdm(testing_set):
                    Interest_points = Gabor(V, 4, 1.5, 0.05)
                    testing_set_desc.append(hist_descriptor(V, Interest_points, descriptor, 8))
                
        bow_train, bow_test = p3.bag_of_words(np.asarray(training_set_desc), np.asarray(testing_set_desc), 64)
        accuracy, pred = p3.svm_train_test(bow_train, train_labels, bow_test, test_labels)
        
        print("Interest point detector: {} and Descriptor: {} achieve accuracy: ".format(detector, descriptor) + str(round(accuracy,2)))
```

```python
# Step 1: Detect SIFT interest points and compute their descriptors  

def SIFT_detect(img1):
    '''
    Εξάγει τα features με τον SIFT για μια εικόνα και τους περιγραφητές τους.
    
    '''
    sift = cv2.xfeatures2d.SIFT_create()
    
    (kp_1, desc_1) = sift.detectAndCompute(img1,None)
    
    return kp_1, desc_1

kp_1, desc_1 = SIFT_detect(Img1)
kp_2, desc_2 = SIFT_detect(Img2)

Img_feats_1 = cv2.drawKeypoints(Img1, kp_1, 0)
Img_feats_2 = cv2.drawKeypoints(Img2, kp_2, 0)

# print(kp_1[0].pt)

print("Num features img1: " + str(len(kp_1)))
print("Descriptor size: " + str(desc_1[0].shape))

plt.imshow(Img_feats_1)
plt.show()

print("Num features img2: " + str(len(kp_2)))
print("Descriptor size: " + str(desc_2[0].shape))

plt.imshow(Img_feats_2)
plt.show()
```

```python
# Step 2-3: Return Best Matches of keypoints in img1 and keypoints in img2. 

def FLANN_based_matching(img1, kp1, desc1, img2, kp2, desc2):
    '''
    Ματσάρει κάποια features της img1 καθένα με κάποιο feature της img2
    
    '''
    # FLANN parameters
    FLANN_INDEX_KDTREE = 0
    index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)
    search_params = {}
    
    flann = cv2.FlannBasedMatcher(index_params,search_params)
    matches = flann.knnMatch(desc1, desc2, k=2)
    
    # selecting only good matches.
    # matchesMask = [[0,0] for i in range(len(matches))]
    
    good = []
    for i,(m,n) in enumerate(matches):
        if( m.distance < 0.8*n.distance):
            # matchesMask[i]=[1,0]
            good.append(m)
    
    # draw_params = dict(matchColor = (0,255,0), singlePointColor = (255,0,0), matchesMask = matchesMask,flags = 0)
    
    # drawing nearest neighbours
    # img = cv2.drawMatchesKnn(img1, kp1, img2, kp2, matches, None, **draw_params)
    
    # plt.imshow(img)
    # plt.show()
    
    # return location of matches
    src_pts = np.float32([ kp1[m.queryIdx].pt for m in good ]).reshape(-1,1,2)
    dst_pts = np.float32([ kp2[m.trainIdx].pt for m in good ]).reshape(-1,1,2)
    
    return src_pts, dst_pts

src_pts, dst_pts = FLANN_based_matching(Img1, kp_1, desc_1, Img2, kp_2, desc_2)

print(dst_pts.shape)

p1 = src_pts[0][0]
p2 = dst_pts[0][0]

print(str(p1) + " point of img1 matches with " + str(p2) + " on img2.")
```

```python
# Step 4: find homography transformation

'''
Με βάση τα features αυτή εδώ η μαγική συνάρτηση μου υπολογίζει ένα πίνακα ομογενών συντεταγμένων που στέλνει ένα σημείο
της img1 στο πλαίσιο αναφοράς της img2.

'''

H, Mask = cv2.findHomography(src_pts, dst_pts, cv2.RANSAC,5.0)

print(H)
```

```python
# Step 5 warp image

def projectionImage(H, Img1):
    '''
    "Μεταφράζω" την img1 στο πλαίσιο αναφοράς της img2.
    
    '''
    h,w = Img1.shape[:2]
    pts = np.float32([ [0,0],[0,h-1],[w-1,h-1],[w-1,0] ]).reshape(-1,1,2)

    dst = cv2.perspectiveTransform(pts,H)
    
    if dst[0][0][0] < 0:

        img1_topleft_coords = dst[0][0]
        pt_A = dst[0][0]
        pt_B = dst[1][0]
        pt_C = dst[2][0]
        pt_D = dst[3][0]

        width_AD = np.sqrt(((pt_A[0] - pt_D[0]) ** 2) + ((pt_A[1] - pt_D[1]) ** 2))
        width_BC = np.sqrt(((pt_B[0] - pt_C[0]) ** 2) + ((pt_B[1] - pt_C[1]) ** 2))
        maxWidth = max(int(width_AD), int(width_BC))
 
        height_AB = np.sqrt(((pt_A[0] - pt_B[0]) ** 2) + ((pt_A[1] - pt_B[1]) ** 2))
        height_CD = np.sqrt(((pt_C[0] - pt_D[0]) ** 2) + ((pt_C[1] - pt_D[1]) ** 2))
        maxHeight = max(int(height_AB), int(height_CD))

        T = np.array([[1, 0, 0],
                      [0, 1, 0],
                      [0, 0, 1]], dtype = 'float32')
        T[0][2] = -img1_topleft_coords[0]
        T[1][2] = -img1_topleft_coords[1]

        Img1_warped = cv2.warpPerspective(Img1,T@H,(maxWidth, maxHeight),flags=cv2.INTER_LINEAR)
    
        return Img1_warped, img1_topleft_coords
    
    else:
        
        img1_topleft_coords = dst[3][0]
        pt_A = dst[0][0]
        pt_B = dst[1][0]
        pt_C = dst[2][0]
        pt_D = dst[3][0]

        width_AD = np.sqrt(((pt_A[0] - pt_D[0]) ** 2) + ((pt_A[1] - pt_D[1]) ** 2))
        width_BC = np.sqrt(((pt_B[0] - pt_C[0]) ** 2) + ((pt_B[1] - pt_C[1]) ** 2))
        maxWidth = max(int(width_AD), int(width_BC)) + int(np.ceil(dst[0][0][0]))
 
        height_AB = np.sqrt(((pt_A[0] - pt_B[0]) ** 2) + ((pt_A[1] - pt_B[1]) ** 2))
        height_CD = np.sqrt(((pt_C[0] - pt_D[0]) ** 2) + ((pt_C[1] - pt_D[1]) ** 2))
        maxHeight = max(int(height_AB), int(height_CD)) + int(np.ceil(dst[0][0][1]))

        T = np.array([[1, 0, 0],
                      [0, 1, 0],
                      [0, 0, 1]], dtype = 'float32')
        T[0][2] = -dst[3][0][0] + maxWidth
        T[1][2] = -dst[3][0][1]

        Img1_warped = cv2.warpPerspective(Img1,T@H,(maxWidth, maxHeight),flags=cv2.INTER_LINEAR)
    
        return Img1_warped, img1_topleft_coords

Img1_warped, img1_topleft_coords = projectionImage(H, Img1)
    
plt.imshow(Img1_warped)
plt.show()
```

```python
# Step 6: Stitching

def mergeWarpedImages(Img1_warped, Img2, img1_topleft_coords):
    '''
    Με χειρουργικές κινήσεις κολλάω τις δύο εικόνες.
    
    '''
    
    if img1_topleft_coords[0] < 0:

        h,w = Img2.shape[:2]

        maxWidth = w-np.ceil(img1_topleft_coords[0])
        maxHeight = h-np.ceil(img1_topleft_coords[1])
    
        T = np.array([[1, 0, 0],
                      [0, 1, 0],
                      [0, 0, 1]], dtype = 'float32')
        T[0][2] = -img1_topleft_coords[0] 
        T[1][2] = -img1_topleft_coords[1]

        Img2_translated = cv2.warpPerspective(Img2,T,(int(maxWidth), int(maxHeight)),flags=cv2.INTER_LINEAR)
    
        # extend both images to have same shape
        I = np.array([[1, 0, 0],
                      [0, 1, 0],
                      [0, 0, 1]], dtype = 'float32')

        maxWidth = max(Img1_warped.shape[1], Img2_translated.shape[1])
        maxHeight = max(Img1_warped.shape[0], Img2_translated.shape[0])

        Img1_warped = cv2.warpPerspective(Img1_warped,I,(maxWidth, maxHeight),flags=cv2.INTER_LINEAR)
        Img2_translated = cv2.warpPerspective(Img2_translated,I,(maxWidth, maxHeight),flags=cv2.INTER_LINEAR)
    
        # Add them up
        Nonzero = Img2_translated == 0

        I = np.where(Nonzero, Img1_warped, 0)

        return I + Img2_translated
    
    else:
        
        h,w = Img2.shape[:2]
        width = Img1_warped.shape[1]

        maxWidth = w
        maxHeight = h-np.ceil(img1_topleft_coords[1])
    
        T = np.array([[1, 0, 0],
                      [0, 1, 0],
                      [0, 0, 1]], dtype = 'float32')
        T[0][2] = -img1_topleft_coords[0] + width
        T[1][2] = -img1_topleft_coords[1]

        Img2_translated = cv2.warpPerspective(Img2,T,(int(maxWidth), int(maxHeight)),flags=cv2.INTER_LINEAR)
    
        # extend both images to have same shape
        I = np.array([[1, 0, 0],
                      [0, 1, 0],
                      [0, 0, 1]], dtype = 'float32')

        maxWidth = max(Img1_warped.shape[1], Img2_translated.shape[1])
        maxHeight = max(Img1_warped.shape[0], Img2_translated.shape[0])

        Img1_warped = cv2.warpPerspective(Img1_warped,I,(maxWidth, maxHeight),flags=cv2.INTER_LINEAR)
        Img2_translated = cv2.warpPerspective(Img2_translated,I,(maxWidth, maxHeight),flags=cv2.INTER_LINEAR)
    
        # Add them up
        Nonzero = Img2_translated == 0

        I = np.where(Nonzero, Img1_warped, 0)

        return I + Img2_translated


stitched_image = mergeWarpedImages(Img1_warped, Img2, img1_topleft_coords)

plt.imshow(stitched_image)
plt.show()
```

```python
# End to end stitching algorithm

def stitchImages(Img1, Img2):
    # detect features and compute their descriptors
    kp_1, desc_1 = SIFT_detect(Img1)
    kp_2, desc_2 = SIFT_detect(Img2)
    
    # Match features of Img1 with features on Img2 
    src_pts, dst_pts = FLANN_based_matching(Img1, kp_1, desc_1, Img2, kp_2, desc_2)
    
    # Compute Homography transformation H : Img1 ----> Img2
    H, Mask = cv2.findHomography(src_pts, dst_pts, cv2.RANSAC,10.0)
    
    # Warp Img1 according to H
    Img1_warped, img1_topleft_coords = projectionImage(H, Img1)
    
    # print(Img1_warped.shape[1])
    # print(img1_topleft_coords)
    
    # Stitch images
    stitched_image = mergeWarpedImages(Img1_warped, Img2, img1_topleft_coords)
    
    return stitched_image

P_1_2_3 = stitchImages(stitchImages(Img1,Img2),Img3)

plt.imshow(P_1_2_3)
plt.show()

P_3_4_5 = stitchImages(stitchImages(Img5,Img4),Img3)

plt.imshow(P_3_4_5)
plt.show()


# extend both images to have same shape
for i in range(3):
    for x in range(Img3.shape[1] + 1027 - 961):
        for y in range(Img3.shape[0] + 384):
            P_3_4_5[y,x,i] = 0.0
            
T = np.array([[1, 0, 0],
              [0, 1, 0],
              [0, 0, 1]], dtype = 'float32')

T[0][2] = 943.26337 - 1027 + 961.45636
T[1][2] = 640.1014 - 383.97153

maxWidth = int(max(np.ceil(T[0][2]) + P_3_4_5.shape[1], P_1_2_3.shape[1])) 
maxHeight = int(max(np.ceil(T[1][2]) + P_3_4_5.shape[0], P_1_2_3.shape[0])) 

P_3_4_5 = cv2.warpPerspective(P_3_4_5,T,(maxWidth, maxHeight),flags=cv2.INTER_LINEAR)

I = np.array([[1, 0, 0],
              [0, 1, 0],
              [0, 0, 1]], dtype = 'float32')

P_1_2_3 = cv2.warpPerspective(P_1_2_3,I,(maxWidth, maxHeight),flags=cv2.INTER_LINEAR)
    
# Add them up
Panorama = P_1_2_3 + P_3_4_5

plt.imshow(Panorama)
plt.show()
```
